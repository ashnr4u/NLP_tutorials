```python
# Separate the input (Hungarian) and target (English) sentences into separate lists.
SEPARATOR = '<sep>'

result=[]
for pair in train:
  result.append(pair.split(SEPARATOR))
zipped = zip(*result)

'''

Before (rows):              After (columns):

[hun1 , eng1]               (hun1, hun2, hun3)
[hun2 , eng2]     ───►      (eng1, eng2, eng3)
[hun3 , eng3]

'''


train_input, train_target = map(list,zipped)
```
