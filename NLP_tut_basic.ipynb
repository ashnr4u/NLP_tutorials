```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
class Category:
  BOOKS= "BOOKS"
  CLOTHING=" CLOTHING"
```

```python
train_x= ["i love this book book","this is a great book",
          "this fit is great","i love the shoes"]
train_y= [Category.BOOKS,Category.BOOKS,Category.CLOTHING,Category.CLOTHING]
```

```python
vectorizer =CountVectorizer(binary =True) # ngram_range=(1,2) this will make two pair of words
train_x_vectors = vectorizer.fit_transform(train_x) #we transform the training data set using this...the words get converted into vectors
print(vectorizer.get_feature_names_out())
print(train_x_vectors)
```

```python
from sklearn import svm
clf_svm= svm.SVC(kernel='linear')
clf_svm.fit(train_x_vectors,train_y)
```

```python
test_x= vectorizer.transform(['this is my bookstore'])
```

```python
clf_svm.predict(test_x)
```


# Word Vectors
> In this approach, we use spaCy to convert each sentence into a dense vector representation that captures semantic meaning. These vectors are then used as input to an SVM classifier. This method improves generalization compared to bag-of-words because similar words have similar vector representations, allowing the model to understand meaning rather than just word frequency.

```python
import spacy
```

```python
print(train_x)
```

```python
nlp=spacy.load("en_core_web_md")
```

```python
docs= [nlp(text) for text in train_x]
train_x_word_vectors = [x.vector for x in docs]
```

```python
# docs[0].vector vector represtantion of the first sentence
```

```python
from sklearn import svm
clf_svm_wv = svm.SVC(kernel="linear")
clf_svm_wv.fit(train_x_word_vectors,train_y)
```



> “In bag-of-words models, predictions are limited to the vocabulary seen during training. However, when using spaCy word vectors, each word has a pretrained embedding, so even unseen words can be represented meaningfully. This allows the model to generalize better to new or similar words.”



```python
test_x= ["i love the books"]
test_docs = [nlp(text) for text in test_x]
```

```python
test_x_word_vectors = [x.vector for x in test_docs]
clf_svm_wv.predict(test_x_word_vectors)
```

# STEMMING AND LEMITIZATION
When to use what
Use stemming when:
speed matters (e.g., search engines, large-scale text processing)
Use lemmatization when:
meaning matters (e.g., NLP models, chatbots, analysis)

```python
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```



*  Stemming A rough, rule-based chopping of word endings. It doesn’t care about actual meaning or grammar.
Result may not be a real word.
*   Lemmatization
Uses vocabulary + grammar rules (and often POS tagging).

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer =PorterStemmer()
phrase= "The striped bats are hanging on their feet and ate best fishes."
words= word_tokenize(phrase)
stemmed_words = []
for word in words:
  stemmed_words.append(stemmer.stem(word))
" ".join(stemmed_words)
```

```python
#lemmatizing
from nltk.stem import  WordNetLemmatizer
lemmatizer= WordNetLemmatizer()

phrase= "The striped bats are hanging on their feet and ate best fishes."
words= word_tokenize(phrase)
# print("words after tokenization",words)
lemmatized_words=[]

for word in words:
  lemmatized_words.append(lemmatizer.lemmatize(word,pos='v'))

" ".join(lemmatized_words)
```

# STOPWORDS
It removes common, low-meaning words (stopwords) from a sentence to keep only important words.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# One-time download of the stopwords corpus

stop_words= stopwords.words('english')
stop_words.remove("here")
phrase='here is an example sentence demonstrating the removal of stopwords'
words= word_tokenize(phrase)
stripped_phrase = [word for word in words if word not in stop_words]
# stripped_phrase=[]
# for word in words:
#   if word not in stop_words:
#     stripped_phrase.append(word)
" ".join(stripped_phrase)
```
